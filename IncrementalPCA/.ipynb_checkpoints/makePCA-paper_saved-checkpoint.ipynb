```python


def projectData(data,ipca_file):
    
    #load the ipca
    ipca = pk.load(open(ipca_file,'rb'))
    #print("explained variance : \n"+str(ipca.explained_variance_))
  
    batchsize=30000
    batch_projections=[]
    for i in range(round(data.shape[0]/batchsize+0.5)):
        (start,end)=(i*batchsize,(i+1)*batchsize)
        print("project "+str(start)+":"+str(end), end = '')
        batch_projections.append(ipca.transform((data)[start:end]))
        print("->ok")
    full_projection=np.concatenate(batch_projections, axis=0)
  
    
    return(full_projection,ipca.explained_variance_)
```

```python

datafile="/home/poujolra/projects/rrg-hussinju/shared/covid-19/database/data/inter/2021_01_20_imputed_hapV3/dataformat/working_on_pca/paper_final_Alpha.data"
varfile="/home/poujolra/projects/rrg-hussinju/shared/covid-19/database/data/inter/2021_01_20_imputed_hapV3/dataformat/working_on_pca/paper_final_Alpha.var"

minAlt=10

saved_snpnames=datafile+".snpNames_"+str(minAlt)+"_noref.pkl"
saved_df=datafile+".df_"+str(minAlt)+"_noref.pkl"

########################################
########################################
########  SELECTION OF SNPS ############
########################################
########################################

def select_SNP(datafile,varfile,minAlt):
    listnuc=["A","C","G","T"]

    var=pd.read_csv(varfile, delim_whitespace=True)
    var=var[var['POS']>55]
    var=var[var['POS']<29854]
    n_sample=int(var[1:2]["A"])+int(var[1:2]["C"])+int(var[1:2]["G"])+int(var[1:2]["T"])+int(var[1:2]["."])
    too_much_missing=str(len(var[var['.']>int(n_sample/10)]))

    print(too_much_missing+ " sites where remove because of more than "+str(n_sample/10)+" missing data (10%)")
    var=var[var['.']<=int(n_sample/10)]

    snp_names=[]
    for ref in listnuc :
        alt_listnuc=[a for a in listnuc if a!=ref]
        for alt in alt_listnuc:
            d=list(var.loc[(var['REF']==ref) & (var[alt]> minAlt) & (var[alt]+var["."]!= n_sample)]["POS"])
            snp_names=snp_names+[str(i)+"_"+ref+"_"+alt for i in d]
    nb_snp=len(snp_names)
    print(str(nb_snp)+" SNP kept with at least "+str(minAlt)+" alternative allele")
    snp_names=sorted(snp_names, key=lambda i: int(i.split("_")[0]))
    snplist=[(int(i.split("_")[0]),i.split("_")[2]) for i in snp_names]
    return snplist
```

```python
def getposfromlist(row,cn):
    for c in cn :
        if row[c]: return c[-1]
    return cn[0][-3]

def gethap(s): return(Hap_dict.get(s,"NA"))

#Return the reference state of a given position
def getposref(pos):
    return refseq[pos]

#Return the state of a given position for a list of sequence (rows)
#Compute the subset of column names with the given position
def getposdata(dat,pos):
    pp="^"+str(pos)+"_"
    collist=list(dat.filter(regex=pp,axis=1).columns)
    if collist==[] : print("Cannot retrieve hap : no information for "+pp)
    return(dat.apply (lambda row: getposfromlist(row,collist), axis=1))

#Return a list of haplotype sequence from a matrix and a list of position
def gethapseqdata(dat):
    a=""
    for p in range(len(Hap_posList)):
        a+=getposdata(dat,Hap_posList[p])
    return(a)


#Return a list of haplotype name from a matrix and a list of position
def gethapdata(dat):
    return([gethap(i) for  i in gethapseqdata(dat)])

gethapdata(data)
```
